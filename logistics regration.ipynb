```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
ds=pd.read_csv(r"C:\Users\dell\Downloads\Titanic (1).csv")
ds
```

```python
ds.isnull().sum()
```

```python
ds['Age'].fillna(ds['Age'].median(),inplace=True)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
ds['Embarked'].fillna(ds['Embarked'].mode(),inplace=True)
```

```python
ds['Embarked']=LabelEncoder().fit_transform(ds['Embarked'])
```

```python
ds['Sex']=LabelEncoder().fit_transform(ds['Sex'])
```

```python
ds
```

```python
x=ds[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
x
```

```python
y=ds[['Survived']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=1000) 
model.fit(x_train,y_train)
```

```python
import joblib
joblib.dump(model,'titanic_model')
```

* confusion_matrix elements

1. True positive(TP):
    * These are cases where the model correctly predicted the positive class
2. False positive(FP):
    *  These are cases where the model incorrectly predicted the positive class
3. True negative(TN):
    *  These are cases where the model correctly predicted the negative class
4. False negative(FN):
    *  These are cases where the model incorrectly predicted the negative class
